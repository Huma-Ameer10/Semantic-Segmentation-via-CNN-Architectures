**SPILITTING DATA**

```python
import os
import numpy as np
import shutil
#for Deviding into test and train set
root_dir = "/content/A3_Dataset/"
val_ratio = 0.20

os.makedirs(root_dir +'annotations_train/')
os.makedirs(root_dir +'annotations_val/')
os.makedirs(root_dir +'images_train/')
os.makedirs(root_dir +'images_val/')

src = root_dir + 'images_prepped_train'
dataFiles = os.listdir(src)
np.random.shuffle(dataFiles)

train_dataFile, val_dataFile = np.split(np.array(dataFiles),[int(len(dataFiles)* (1 - val_ratio))])
train_dataFile = [ name for name in train_dataFile.tolist()]
val_dataFile   = [ name for name in val_dataFile.tolist()]

print("***********")
print('Total_images: ', len(dataFiles))
print('Training_data: ', len(train_dataFile))
print('Validation-data: ', len(val_dataFile))
print("***********")

for name in train_dataFile:
      source_img = root_dir + 'images_prepped_train/' + name
      dest_img=root_dir + 'images_train' 
      shutil.copy(source_img,dest_img)
      source_ann = root_dir + 'annotations_prepped_train/' + name
      dest_ann=root_dir + 'annotations_train'
      shutil.copy(source_ann,dest_ann)
print("Copying train data: Done!")        
for name in val_dataFile:
      source_img = root_dir + 'images_prepped_train/' + name
      dest_img=root_dir + 'images_val' 
      shutil.copy(source_img,dest_img)
      source_ann = root_dir + 'annotations_prepped_train/' + name
      dest_ann=root_dir + 'annotations_val'
      shutil.copy(source_ann,dest_ann)
print("Copying validation data: Done!")        
```

**MODEL FITTING**

```python

from keras_segmentation.models.unet import vgg_unet

model = vgg_unet(n_classes=12 ,  input_height=384, input_width=480 )

```

```python
r=model.train(
    train_images =  "/content/A3_Dataset/images_train",
    train_annotations = "/content/A3_Dataset/annotations_train",
    input_height=384,
    input_width=480,
    n_classes=12,
    do_augment=True,
    val_images="/content/A3_Dataset/images_val/",
    val_annotations="/content/A3_Dataset/annotations_val/",
    augmentation_name="aug_all",
    epochs=50,
    validate=True,
    batch_size=2,
    val_batch_size=2,
    steps_per_epoch=146,
    val_steps_per_epoch=37,
)
```
